**Creating a Small Dashboard to Visualize the Network Logs**

```python
import pandas as pd
import matplotlib.pyplot as plt


# 1. Create sample network_logs

data = {
    "log_id":        [1, 2, 3, 4, 5, 6, 7, 8],
    "timestamp":     [
        "2025-11-30 09:00:01",
        "2025-11-30 09:00:05",
        "2025-11-30 09:00:10",
        "2025-11-30 09:00:15",
        "2025-11-30 09:00:20",
        "2025-11-30 09:00:25",
        "2025-11-30 09:00:30",
        "2025-11-30 09:00:35",
    ],
    "source_ip":     ["10.0.0.5", "10.0.0.12", "192.168.1.20", "172.16.0.4",
                      "10.0.0.5", "192.168.1.30", "10.0.0.12", "172.16.0.4"],
    "destination_ip":["8.8.8.8", "10.1.2.3", "8.8.4.4", "10.1.2.3",
                      "1.1.1.1", "10.1.2.3", "8.8.8.8", "203.0.113.5"],
    "protocol":      ["HTTPS", "HTTPS", "DNS", "SSH",
                      "HTTP", "HTTPS", "SSH", "DNS"],
    "bytes_sent":    [520, 2100, 80, 9500, 320, 1800, 11200, 120],
    "bytes_received":[1020, 3300, 140, 2000, 450, 2600, 500, 220],
    "country":       ["IN", "US", "IN", "RU", "IN", "DE", "US", "CN"],
    "threat_score":  [5, 15, 2, 80, 10, 25, 90, 60]
}

network_logs = pd.DataFrame(data)

# 2. Feature engineering for the dashboard

# Total bytes
network_logs["total_bytes"] = (
    network_logs["bytes_sent"] + network_logs["bytes_received"]
)

# Suspicious flag
suspicious_countries = ["RU", "CN"]
network_logs["is_suspicious"] = (
    (network_logs["threat_score"] >= 70) |
    (network_logs["bytes_sent"] > 10000) |
    (network_logs["country"].isin(suspicious_countries))
)

# Optional: threat level label
def classify_threat(score):
    if score >= 70:
        return "High"
    elif score >= 30:
        return "Medium"
    else:
        return "Low"

network_logs["threat_level"] = network_logs["threat_score"].apply(classify_threat)

display(network_logs)

# 3. Aggregations for dashboard widgets

# Top suspicious IPs by total threat score
top_ips = (
    network_logs[network_logs["is_suspicious"] == True]
    .groupby("source_ip")["threat_score"]
    .sum()
    .reset_index()
    .sort_values(by="threat_score", ascending=False)
)

# Threat score distribution data
threat_scores = network_logs["threat_score"]

# Average threat by country
avg_threat_country = (
    network_logs
    .groupby("country")["threat_score"]
    .mean()
    .reset_index()
    .sort_values(by="threat_score", ascending=False)
)

# 4. Plotting the Threat Dashboard

plt.figure(figsize=(14, 10))

# ---- (1) Top Suspicious IPs ----
plt.subplot(2, 2, 1)
plt.bar(top_ips["source_ip"], top_ips["threat_score"])
plt.title("Top Suspicious Source IPs")
plt.xlabel("Source IP")
plt.ylabel("Total Threat Score")
plt.grid(axis="y", linestyle="--", alpha=0.5)

# ---- (2) Threat Score Distribution ----
plt.subplot(2, 2, 2)
plt.hist(threat_scores, bins=5)
plt.title("Threat Score Distribution")
plt.xlabel("Threat Score")
plt.ylabel("Frequency")
plt.grid(axis="y", linestyle="--", alpha=0.5)

# ---- (3) Average Threat Score per Country ----
plt.subplot(2, 2, 3)
plt.bar(avg_threat_country["country"], avg_threat_country["threat_score"])
plt.title("Average Threat Score by Country")
plt.xlabel("Country")
plt.ylabel("Average Threat Score")
plt.grid(axis="y", linestyle="--", alpha=0.5)

# ---- (4) Bytes Sent vs Bytes Received ----
plt.subplot(2, 2, 4)
plt.scatter(network_logs["bytes_sent"], network_logs["bytes_received"])
plt.title("Bytes Sent vs Bytes Received")
plt.xlabel("Bytes Sent")
plt.ylabel("Bytes Received")
plt.grid(True, linestyle="--", alpha=0.5)

plt.suptitle("Threat Monitoring Dashboard", fontsize=16, fontweight="bold")
plt.tight_layout(rect=[0, 0.03, 1, 0.96])
plt.show()
```

# **Creating Calculator using Python in 3 Ways**

### **1A. Very simple calculator**

```python
# Basic calculator with hard-coded numbers

a = 10      # first number
b = 5       # second number

addition = a + b
subtraction = a - b
multiplication = a * b
division = a / b

print("a =", a, ", b =", b)
print("Addition:", addition)
print("Subtraction:", subtraction)
print("Multiplication:", multiplication)
print("Division:", division)
```

### **1B. Interactive calculator using input() and a function**

```python
def calculator():
    print("Simple Calculator")
    print("Operations: +  -  *  /")

    # take inputs from user (as strings) and convert to float
    num1 = float(input("Enter first number: "))
    op   = input("Enter operator (+, -, *, /): ")
    num2 = float(input("Enter second number: "))

    if op == "+":
        result = num1 + num2
    elif op == "-":
        result = num1 - num2
    elif op == "*":
        result = num1 * num2
    elif op == "/":
        if num2 == 0:
            print("Error: Division by zero is not allowed.")
            return
        result = num1 / num2
    else:
        print("Invalid operator")
        return

    print("Result:", result)

# call the function
calculator()
```

### **1C. Reusable calculator that keeps running until user exits**

```python
def calculator_loop():
    print("Calculator (type 'exit' to quit)")

    while True:
        op = input("\nEnter operator (+, -, *, /) or 'exit': ")
        if op == "exit":
            print("Goodbye!")
            break

        if op not in ["+", "-", "*", "/"]:
            print("Invalid operator, try again.")
            continue

        num1 = float(input("Enter first number: "))
        num2 = float(input("Enter second number: "))

        if op == "+":
            result = num1 + num2
        elif op == "-":
            result = num1 - num2
        elif op == "*":
            result = num1 * num2
        elif op == "/":
            if num2 == 0:
                print("Error: Division by zero.")
                continue
            result = num1 / num2

        print("Result:", result)

calculator_loop()
```
