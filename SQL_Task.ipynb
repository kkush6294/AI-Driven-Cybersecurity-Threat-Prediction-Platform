```python
import sqlite3

# Create or open database
conn = sqlite3.connect('/content/mydatabase.db')
cur = conn.cursor()

# Create Students table
cur.execute("""
CREATE TABLE IF NOT EXISTS students (
    student_id INTEGER PRIMARY KEY,
    name TEXT,
    location TEXT,
    course TEXT,
    score INTEGER
);
""")

conn.commit()
print("Students table created successfully.")
```

```python
import os
import sqlite3

db_path = '/content/mydatabase.db'

# Remove existing database file if it exists (to clear corruption)
if os.path.exists(db_path):
    os.remove(db_path)
    print(f"Removed existing database file: {db_path}")

# Create or open database connection
conn = sqlite3.connect(db_path)
cur = conn.cursor()

# Re-create Students table
cur.execute("""
CREATE TABLE IF NOT EXISTS students (
    student_id INTEGER PRIMARY KEY,
    name TEXT,
    location TEXT,
    course TEXT,
    score INTEGER
);
""")

# Re-insert students data
students_data = [
    (1, 'Govind', 'Andheri', 'BTech', 88),
    (2, 'Gajanan', 'Borivali', 'BSc IT', 72),
    (3, 'Jaydeep', 'Kandivali', 'Diploma CS', 67),
    (4, 'Adharsh', 'Malad', 'BTech', 91),
    (5, 'Aniket', 'Dadar', 'BMS', 75),
    (6, 'Vedansh', 'Bandra', 'BCom', 82),
    (7, 'Ishaan', 'Goregaon', 'BSc IT', 79),
    (8, 'Sahil', 'Jogeshwari', 'BTech', 85),
    (9, 'Nishank', 'Mira Road', 'BCA', 73)
]
cur.executemany("INSERT INTO students VALUES (?, ?, ?, ?, ?)", students_data)

# Re-create student_details table (from the failing cell)
cur.execute("""
CREATE TABLE IF NOT EXISTS student_details (
    id INTEGER PRIMARY KEY,
    full_name TEXT,
    dob DATE,
    email TEXT,
    percentage REAL,
    is_active BOOLEAN
);
""")

conn.commit()
print("Database reset and tables created successfully.")
```

```python
cur.execute("""
CREATE TABLE IF NOT EXISTS courses (
    course TEXT PRIMARY KEY,
    course_name TEXT,
    duration_months INTEGER
);
""")

courses_data = [
    ('BTech', 'Bachelor of Technology', 48),
    ('BSc IT', 'Bachelor of Science (IT)', 36),
    ('BCom', 'Bachelor of Commerce', 36),
    ('BMS', 'Bachelor of Management Studies', 36),
    ('BCA', 'Bachelor of Computer Applications', 36),
    ('Diploma CS', 'Diploma in Computer Science', 24)
]

cur.executemany("INSERT INTO courses VALUES (?, ?, ?)", courses_data)
conn.commit()

print("courses table created and data inserted.")
```

```python
pd.read_sql_query("""
SELECT s.name, s.course, c.course_name, c.duration_months
FROM students s
INNER JOIN courses c
ON s.course = c.course;
""", conn)
```
